```python
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
sns.set()
```

plotting a vector

```python
plt.quiver(0,0,4,5, scale_units = 'xy', angles = 'xy', scale = 1)
plt.xlim(-8, +8)
plt.ylim(-8, +8)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units = 'xy', angles = 'xy', scale = 1, color= 'b')
plt.quiver(0,0,-3,-5, scale_units = 'xy', angles = 'xy', scale = 1, color= 'y')
plt.xlim(-8, +8)
plt.ylim(-8, +8)
plt.show()
```

addition of two vectors

```python
vec_1 = np.asarray([0,0,2,3])
vec_2 = np.asarray([0,0,3,-2])

sum = vec_1 +vec_2
print(sum)
```

```python
plt.quiver(0,0,2,3, scale_units = 'xy', angles = 'xy', scale = 1, color= 'b')
plt.quiver(0,0,3,-2, scale_units = 'xy', angles = 'xy', scale = 1, color= 'y')
plt.quiver(0,0,5,1, scale_units = 'xy', angles = 'xy', scale = 1, color= 'r')
plt.xlim(-8, +8)
plt.ylim(-8, +8)
plt.show()
```

```python
vec_1 = np.asarray([0,0,2,3])
vec_2 = np.asarray([0,0,3,-2])

dif = vec_1 - vec_2
print(dif)
```

```python
plt.quiver(0,0,2,3, scale_units = 'xy', angles = 'xy', scale = 1, color= 'b')
plt.quiver(0,0,3,-2, scale_units = 'xy', angles = 'xy', scale = 1, color= 'y')
plt.quiver(0,0,-1,5, scale_units = 'xy', angles = 'xy', scale = 1, color= 'r')
plt.xlim(-8, +8)
plt.ylim(-8, +8)
plt.show()
```

multiplying a vector by a scalar

```python
vec_1 = np.asarray([0,0,2,3])
vec_2 = 2*vec_1
print(vec_2)
```

```python
plt.quiver(0,0,4,6, scale_units = 'xy', angles = 'xy', scale = 1, color = 'y')
plt.quiver(0,0,2,3, scale_units = 'xy', angles = 'xy', scale = 1, color= 'r')
plt.xlim(-8, +8)
plt.ylim(-8, +8)
plt.show()
```

CROSS PRODUCT

```python
a = np.array([2,3])

b = np.array([4,4])

a_cross_b = np.cross(a,b)
```

```python
print(a_cross_b)
```

Projection of 'a' on 'v'

```python
a = np.array([2,5])

v = np.array([8, -6])

proj = (np.dot(a,v)/np.dot(v,v)) * v
```
